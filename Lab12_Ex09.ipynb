# Lab 12 Class Inheritance

## Overriding base class member (Methods)

การสร้างสมาชิกใหม่ขึ้นใน Derived class โดยใช้ signature เดียวกับสมาชิกใน Base class เรียกว่าการ overriding 

สามารถ Overriding ได้ทั้ง variables และ methods

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex09
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
DerivedClass dc = new DerivedClass();
dc.Greeting();

class BaseClass
{
    public string Hello = "Hello From BaseClass";
    public void Greeting()
    {
        Console.WriteLine("This is method in BaseClass");
        Console.WriteLine(Hello);
    }

}
class DerivedClass : BaseClass
{
    public string Hello = "Hello From DerivedClass";
    public void Greeting()
    {
        Console.WriteLine("This is method in DerivedClass");
        Console.WriteLine(Hello);
    }

}
```

Field Hiding (การซ่อนตัวแปร)

ตัวแปร Hello ของ DerivedClass ซ่อนตัวแปร Hello ของ BaseClass

เมื่อเรียกใช้ Hello ภายใน DerivedClass.Greeting() จะใช้ค่าของ Hello จาก DerivedClass ไม่ใช่ของ BaseClass

Method Hiding (การซ่อนเมทอด)

DerivedClass มีเมทอด Greeting() ที่ชื่อเหมือน BaseClass ทำให้เมทอดของ BaseClass ถูกซ่อน

เมทอด Greeting() ของ DerivedClass ทำงานแทน และแสดงค่าของ Hello จาก DerivedClass

การแก้ไขที่ควรทำ

ถ้าต้องการให้ DerivedClass.Greeting() ทำงานแทน BaseClass.Greeting() ควรใช้ override แทน new

ถ้าต้องการเรียก Greeting() ของ BaseClass จาก DerivedClass สามารถใช้ base.Greeting()
