## Masking base class member with `new` keyword

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex11
```



2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
DerivedClass dc = new DerivedClass();
System.Console.WriteLine(dc.Hello);

class BaseClass
{
    public string Hello = "Hello From BaseClass";
}
class DerivedClass : BaseClass
{
    new public string Hello = "Hello From DerivedClass";
}
```



3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab12_Ex11
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง

บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3


5.Run project โดยการใช้คำสั่ง

```cmd
dotnet run --project Lab12_Ex11
```

6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง

การซ่อนตัวแปร (Field Hiding): ตัวแปร Hello ใน DerivedClass ซ่อนตัวแปร Hello ใน BaseClass ทำให้เข้าถึงค่าจาก DerivedClass เสมอ

การใช้คำว่า new: การใช้ new ชัดเจนว่าต้องการซ่อนตัวแปรจากคลาสแม่ แต่ไม่จำเป็นต้องใช้ก็ได้

ผลลัพธ์: เมื่อเรียก dc.Hello จะได้ "Hello From DerivedClass"

การเข้าถึงตัวแปรในคลาสแม่: ต้องใช้ base.Hello ใน DerivedClass เพื่อเข้าถึงตัวแปรจาก BaseClass
