# Lab 12 Class Inheritance

## Access overriding base class member (Methods)

การเข้าถึงสมาชิกของ Base class ที่ถูกปิดบัง ทำได้โดยการแปลงชนิดข้อมูลไปยัง BaseClass เสียก่อน แล้วจึงเข้าถึงสมาชิกที่สืบทอดมาจาก base class นั้น 

สามารถทำได้ 2 แบบ คือ  

1. ใช้วิธีการแปลงประเภท (Type Casting) หรือ การแปลงชนิดข้อมูล (Data Type Conversion)

2. ใช้ keyword `'as'`

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
// 1. ใช้วิธีการแปลงประเภท (Type Casting) หรือ การแปลงชนิดข้อมูล (Data Type Conversion)
DerivedClass dc = new DerivedClass();
Console.WriteLine("Use Type Casting or Data Type Conversion");
((BaseClass)dc).Greeting(); 

// 2. ใช้ keyword 'as'
Console.WriteLine("------------");
Console.WriteLine("Use as keyword");
(dc as BaseClass).Greeting();

class BaseClass
{
    public string Hello = "Hello From BaseClass";
    public void Greeting()
    {
        Console.WriteLine("This is method in BaseClass");
        Console.WriteLine(Hello);
    }

}
class DerivedClass : BaseClass
{
    public string Hello = "Hello From DerivedClass";
    public void Greeting()
    {
        Console.WriteLine("This is method in DerivedClass");
        Console.WriteLine(Hello);
    }
}
```

Method Hiding

ในโค้ดที่ให้มา DerivedClass มีเมทอด Greeting() ที่ซ้ำกับเมทอดใน BaseClass

เมทอดใน DerivedClass จะ ซ่อน (hide) เมทอดของ BaseClass ซึ่งหมายความว่าหากเรียกเมทอดผ่านการแปลงประเภท (type casting) จะเรียกเมทอดของ BaseClass เสมอ

การใช้ Type Casting

เมื่อใช้การแปลงประเภท (BaseClass)dc หรือ dc as BaseClass เพื่อเรียกเมทอด Greeting() จะทำให้โปรแกรมเรียกเมทอดของ BaseClass เสมอ

นี่เป็นเพราะ Greeting() ใน BaseClass ถูกเรียกใช้งานเมื่อทำการแปลงประเภท แต่ไม่ได้มีการ override

ผลลัพธ์ของการแปลงประเภท

การเรียก ((BaseClass)dc).Greeting(); จะพิมพ์ผลลัพธ์จาก BaseClass

การเรียก (dc as BaseClass).Greeting(); จะให้ผลลัพธ์เหมือนกัน โดยเรียกเมทอดของ BaseClass

ค่าของตัวแปร Hello

ในเมทอด Greeting() ของ BaseClass ค่าที่แสดงออกมาคือ "Hello From BaseClass" แม้ว่าจะมีฟิลด์ Hello ใน DerivedClass ที่ชื่อเดียวกัน
