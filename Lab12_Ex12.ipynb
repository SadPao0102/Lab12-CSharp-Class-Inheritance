# Lab 12 Class Inheritance

## Access base member by `base` keyword

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex12
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```cs
DerivedClass dc = new DerivedClass();
dc.PrintBaseHello();

class BaseClass
{
    public string Hello = "Hello From BaseClass";
}
class DerivedClass : BaseClass
{
    new public string Hello = "Hello From DerivedClass";

    public void PrintBaseHello()
    {
        string BaseHello = base.Hello;       // ใช้งานสมาชิกใน base class โดยคีย์เวิร์ด base
        System.Console.WriteLine(BaseHello);
    }
}

```

3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab12_Ex12
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง



4.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3



5.Run project โดยการใช้คำสั่ง

```cmd
dotnet run --project Lab12_Ex12
```



6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง

การใช้ base: คีย์เวิร์ด base ช่วยให้เข้าถึงสมาชิกของคลาสแม่ (BaseClass) ได้ แม้จะมีตัวแปรที่ชื่อเดียวกันในคลาสลูก (DerivedClass)

Field Hiding: ตัวแปร Hello ใน DerivedClass ซ่อนตัวแปร Hello ใน BaseClass ทำให้เมื่อต้องการใช้ตัวแปรของ BaseClass ต้องใช้ base.Hello

ผลลัพธ์ที่แสดง: การเรียก dc.PrintBaseHello() จะแสดงข้อความ "Hello From BaseClass" ซึ่งยืนยันว่าการเข้าถึงสมาชิกในคลาสแม่ทำได้ถูกต้อง
