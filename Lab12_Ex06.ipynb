# Lab 12 Class Inheritance

## Multiple base class inheritance

````
classDiagram
    class BaseClass1
    class BaseClass2
    class DerivedClass

    BaseClass1 <|-- BaseClass2
    BaseClass2 <|-- DerivedClass
````

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex06
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
DerivedClass dc = new DerivedClass();
class BaseClass1
 {
    public BaseClass1()
    {
        System.Console.WriteLine("This is BaseClass1");
    }
 }
class BaseClass2: BaseClass1
 {
    public BaseClass2()
    {
        System.Console.WriteLine("This is BaseClass2");
    }
 }

class DerivedClass : BaseClass2
{
    public DerivedClass()
    {
        System.Console.WriteLine("This is DerivedClass");
    }
}
```



3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab12_Ex06
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง



4.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3



5.Run project โดยการใช้คำสั่ง

```cmd
dotnet run --project Lab12_Ex06
```



6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง

C# รองรับการสืบทอดแบบหลายระดับ (Multilevel Inheritance)

DerivedClass สืบทอดจาก BaseClass2 ซึ่งสืบทอดจาก BaseClass1

ทำให้ DerivedClass สามารถเข้าถึงคุณสมบัติของ BaseClass1 และ BaseClass2 ได้

ลำดับการเรียก Constructor

เมื่อสร้าง DerivedClass จะมีการเรียก Constructor ของคลาสแม่ก่อนเสมอ

ลำดับการเรียก:

BaseClass1() → "This is BaseClass1"

BaseClass2() → "This is BaseClass2"

DerivedClass() → "This is DerivedClass"

สาเหตุที่เป็นแบบนี้เพราะ C# ต้องการให้คลาสลูกมีโครงสร้างที่ถูกต้องก่อนใช้งาน

โครงสร้างการสืบทอดในแนวลึก (Deep Inheritance Structure)

คลาสที่อยู่ระดับสูงสุด (BaseClass1) จะถูกสร้างก่อนเสมอ

คลาสที่อยู่ระดับล่างสุด (DerivedClass) จะถูกสร้างหลังสุด
