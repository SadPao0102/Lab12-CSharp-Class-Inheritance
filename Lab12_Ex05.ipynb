DerivedClass dc = new DerivedClass();

การสืบทอดคลาสเดียว:

ใน C# ไม่สามารถทำการสืบทอดจากหลายคลาส (multiple inheritance) ได้ ซึ่งในกรณีนี้ DerivedClass ถูกประกาศให้สืบทอดจาก BaseClass1 เท่านั้น และไม่สามารถสืบทอดจาก BaseClass2 ได้พร้อมกัน

การใช้การสืบทอดจากคลาสเดียวทำให้โครงสร้างของคลาสเรียบง่ายและไม่ซับซ้อน

การใช้ Composition:

เนื่องจากไม่สามารถสืบทอดจากหลายคลาสได้ จึงใช้ composition โดยการสร้าง instance ของ BaseClass2 ภายใน DerivedClass

วิธีนี้ช่วยให้ DerivedClass สามารถใช้ฟังก์ชันจาก BaseClass2 โดยไม่ต้องสืบทอดจากมัน

ลำดับการเรียก Constructor:

เมื่อสร้าง instance ของ DerivedClass, constructor ของ BaseClass1 จะถูกเรียกก่อนเสมอ จากนั้นสร้าง instance ของ BaseClass2 ใน constructor ของ DerivedClass

ลำดับการทำงานจึงเป็น:

BaseClass1 Constructor → แสดงข้อความ: "This is BaseClass1"

BaseClass2 Constructor → แสดงข้อความ: "This is BaseClass2"

DerivedClass Constructor → แสดงข้อความ: "This is DerivedClass"

การควบคุมการเข้าถึง:

การสร้าง instance ของ BaseClass2 ใน DerivedClass ช่วยให้การเข้าถึงคุณสมบัติหรือฟังก์ชันของ BaseClass2 เป็นไปได้อย่างมีระเบียบ

## คำถาม

1. ในภาษา C#  เราสามารถสืบทอดสมบัติจาก Base class มากกว่าหนึ่งคลาสได้หรือไม่ 

ไม่สามารถ ในภาษา C# ไม่รองรับการสืบทอดจากหลายคลาส (multiple inheritance) หมายความว่าคลาสไม่สามารถสืบทอดจากมากกว่าหนึ่งคลาสแม่ได้ในเวลาเดียวกัน

2. จากข้อ 1 ถ้าไม่ได้แล้ว ในกรณีที่ต้องการสืบทอดจา่กหลาย Type จะทำได้อย่างไร

Interface
Interface เป็นวิธีที่สามารถใช้เพื่อให้สามารถสืบทอดคุณสมบัติจากหลายๆ Type ได้

คลาสสามารถ implement (ใช้) หลาย interfaces ได้ ทำให้คุณสามารถนำฟังก์ชันจากหลายแหล่งมาใช้ในคลาสเดียวกัน

ใช้ composition คือการสร้างคลาสที่มี instance ของคลาสอื่นๆ อยู่ภายใน

ช่วยให้คุณสามารถรวมฟังก์ชันจากหลายคลาสโดยไม่ต้องสืบทอดจากหลายคลาส
